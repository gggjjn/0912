```python
import seaborn as sns
import matplotlib.pyplot as plt

# 
tips = sns.load_dataset("tips")

# 
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("")
plt.show()

# 
sns.histplot(tips['total_bill'], kde=True)
plt.title("")
plt.show()

# 
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("")
plt.show()
```
