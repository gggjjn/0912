```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x

# 绘制折线图
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('折线图')
plt.show()

# 绘制散点图
plt.figure(figsize=(8, 6))
plt.scatter(x, y3, label='x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('散点图')
plt.show()

# 绘制柱状图
categories = ['Category A', 'Category B', 'Category C']
values = [20, 35, 45]

plt.figure(figsize=(8, 6))
plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('柱状图')
plt.show()
```
